```python
<i class="fa fa-arrow-circle-down" aria-hidden="true"> 请在下方代码块中开始编写代码！</i>
import numpy as np

x = np.array([56, 72, 69, 88, 102, 86, 76, 79, 94, 74])
y = np.array([92, 102, 86, 110, 130, 99, 96, 102, 105, 92])
from matplotlib import pyplot as plt 
%matplotlib inline

plt.scatter(x,y)
plt.xlabel("Area")
plt.ylabel("Price")
def f(x,w0,w1):
    y = w0+w1*x
    return y

def square_loss(x,y,w0,w1):
    loss = sum(np.square(y - (w0+w1 * x)))
    return loss

def w_calculator(x,y):
    n = len(x)
    w1 = (n*sum(x*y) - sum(x)*sum(y))/(n*sum(x**2) - sum(x)**2)
    w0 = (sum(x**2)*sum(y) - sum(x)*sum(x*y))/(n*sum(x**2) - sum(x)**2)
    return w0,w1

w_calculator(x,y)
w0,w1 = w_calculator(x,y)

square_loss(x,y,w0,w1)
x_temp = np.linspace(50,120,100)

plt.scatter(x,y)
plt.plot(x_temp,f(x_temp,w0,w1),'r')
f(150,w0,w1)
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x.reshape(len(x),1),y)

model.intercept_,model.coef_
model.predict([[150]])
def w_matrix(x,y):
    w = (x.T * x).I * x.T * y
    return w

x = np.matrix([[1, 56], [1, 72], [1, 69], [1, 88], [1, 102],
               [1, 86], [1, 76], [1, 79], [1, 94], [1, 74]])
y = np.matrix([92, 102, 86, 110, 130, 99, 96, 102, 105, 92])

w_matrix(x, y.reshape(10, 1))
import pandas as pd

df = pd.read_csv("https://labfile.oss.aliyuncs.com/courses/1081/course-5-boston.csv")
df.head()
features = df[['crim','rm','lstat']]
features.describe()
target = df['medv']

split_num = int(len(features) * 0.7)

X_train = features[:split_num]
Y_train = target[:split_num]

X_test = features[split_num:]
Y_test = target[split_num:]


model = LinearRegression()
model.fit(X_train,Y_train)
model.coef_,model.intercept_
preds = model.predict(X_test)
preds[:10]
np.array(Y_test)[:10]
def mae_value(y_true,y_pred):
    n = len(y_true)
    mae = sum(np.abs(y_true - y_pred)) / n
    return mae
def mse_value(y_true,y_pred):
    n = len(y_true)
    mse = sum(np.square(y_pred - y_true))/n
    return mse
mae = mae_value(Y_test.values,preds)
mse = mse_value(Y_test.values,preds)

print("MAE:",mae)
print("MSE:",mse)
```
